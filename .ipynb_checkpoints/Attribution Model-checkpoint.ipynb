<h2> Markov Chain Attribution Model

```python
#Iterate through columns to change all the int to str and remove any trailing .0
cols=attr.columns
for col in cols:
    attr[col]=attr[col].astype(str)
    attr[col]=attr[col].map(lambda x:str(x)[:-2] if '.' in x else str(x))
```

The Markov chain framework wants the user journeys in a single variable and on the form Channel 1 > Channel 2 > Channel 3 > …, so the next loop creates exactly that

```python
# Create a total path variable
attr['Path'] = ''
for i in attr.index:
    #df.at[i, 'Path'] = 'Start'
    for x in cols:
        attr.at[i, 'Path'] = attr.at[i, 'Path'] + attr.at[i, x] + ' > '
```

Since channel 21 in our dataset is a conversion event, we will separate that channel from the path and create a separate conversion variable holding the number of conversions happening (still only 1 in our user journey level data)


```python
# Split path on conversion (channel 21)
attr['Path'] = attr['Path'].map(lambda x: x.split(' > 21')[0])

# Create conversion value we can sum to get total conversions for each path
attr['Conversion'] = 1
```

```python
attr.head()
```

```python
# Select relevant columns
attr = attr[['Path', 'Conversion']]

# Sum conversions by Path
attr = attr.groupby('Path').sum().reset_index()

# Write DF to CSV to be executed in R
attr.to_csv('Paths.csv', index=False)
```

<h2> Last Touch Attribution

```python
# Last Touch Attribution
attr['Last Touch'] = attr['Path'].map(lambda x: x.split(' > ')[-1])
df_lt = pd.DataFrame()
df_lt['Channel'] = attr['Last Touch']
df_lt['Attribution'] = 'Last Touch'
df_lt['Conversion'] = 1
df_lt = df_lt.groupby(['Channel', 'Attribution']).sum().reset_index()
df_lt
```

<h2> First Touch Attribution

```python
# First Touch Attribution
attr['First Touch'] = attr['Path'].map(lambda x: x.split(' > ')[0])
df_ft = pd.DataFrame()
df_ft['Channel'] = attr['First Touch']
df_ft['Attribution'] = 'First Touch'
df_ft['Conversion'] = 1
df_ft = df_ft.groupby(['Channel', 'Attribution']).sum().reset_index()
df_ft
```

<h2>Linear Attribution
    

```python
channel = []
conversion = []
for i in attr.index:
    for j in attr.at[i, 'Path'].split(' > '):
        channel.append(j)
        conversion.append(1/len(attr.at[i, 'Path'].split(' > ')))
lin_att_df = pd.DataFrame()
lin_att_df['Channel'] = channel
lin_att_df['Attribution'] = 'Linear'
lin_att_df['Conversion'] = conversion
lin_att_df = lin_att_df.groupby(['Channel', 'Attribution']).sum().reset_index()
lin_att_df
```

```python
#Merge a;; the four dataframes including Markov's into one
# Concatenate the four data frames to a single data frame
df_total_attr = pd.concat([df_ft, df_lt, lin_att_df, markov])
df_total_attr['Channel'] = df_total_attr['Channel'].astype(int)
df_total_attr.sort_values(by='Channel', ascending=True, inplace=True)


# Visualize the attributions
sns.set_style("whitegrid")
plt.rc('legend', fontsize=15)
fig, ax = plt.subplots(figsize=(16, 10))
sns.barplot(x='Channel', y='Conversion', hue='Attribution', data=df_total_attr)
plt.title('Different Attribution models and their respective Channel Conversion ',fontsize=16,color='red',fontweight='bold')
plt.show();
```

From the above chart, we observe that majprity of the people start conversions at Channel 10 and end at channel 20. Channel 20 looks like the major player in conversion. None of them start their journey here.
